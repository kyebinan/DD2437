```python
import numpy as np
import matplotlib.pyplot as plt

import datagen as dg 
import perceptron_rule as pr
import delta_rule as dr
from utils import visualize_loss_function
```

```python
classA, classB = dg.generate_data()
dg.visualize_data(classA, classB)
classA.shape
```

```python
# shuffle the samples
data, label = dg.shuffle_data(classA, classB)
data.shape
```

# Classification with a single-layer perceptron and analysis

### 1-Compare perceptron learning with the delta learning rule

```python
epochs = 20
learning_rate = 0.002
# Applying perceptron learning online
W, B, errors = pr.perceptron_learning(data, label, epochs = epochs, learning_rate = learning_rate)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(classA[:, 0], classA[:, 1], color='red', label='Class 1')
plt.scatter(classB[:, 0], classB[:, 1], color='blue', label='Class -1')
x = np.array([np.min(data[:, 0]), np.max(data[:, 0])])
y = -(B[-1] + W[-1][0] * x)/W[-1][1]
plt.plot(x, y)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title(f'perceptron learning - Epoch {epochs}')
plt.legend()
plt.grid(True)
plt.show()

visualize_loss_function(errors=errors, title='perceptron learning - loss function')
```

```python
epochs = 20
learning_rate = 0.001
# Applying perceptron learning online
W, errors = dr.delta_learning_online(data, label, epochs=epochs, learning_rate=learning_rate)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(classA[:, 0], classA[:, 1], color='red', label='Class 1')
plt.scatter(classB[:, 0], classB[:, 1], color='blue', label='Class -1')
x = np.array([np.min(data[:, 0]), np.max(data[:, 0])])
y = -(W[-1][2] + W[-1][0] * x)/W[-1][1]
plt.plot(x, y)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title(f'delta rule - Epoch {epochs}')
plt.legend()
plt.grid(True)
plt.show()

visualize_loss_function(errors=errors, title='delta rule - loss function')
```
