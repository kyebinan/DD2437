```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

import datagen as dg 
import perceptron_rule as pr
```

```python
classA, classB = dg.generate_data()
dg.visualize_data(classA, classB)
classA.shape
```

```python
# shuffle the samples
data, label = dg.shuffle_data(classA, classB)
data.shape
```

# Classification with a single-layer perceptron and analysis

```python
epochs = 25
learning_rate = 0.001
# Applying perceptron learning 
W, B = pr.perceptron_learning(data, label, epochs = epochs, learning_rate = learning_rate)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(classA[:, 0], classA[:, 1], color='red', label='Class 1')
plt.scatter(classB[:, 0], classB[:, 1], color='blue', label='Class -1')
x = np.array([np.min(data[:, 0]), np.max(data[:, 0])])
y = -(B[-1] + W[-1][0] * x)/W[-1][1]
plt.plot(x, y)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title(f'perceptron learning - Epoch {epochs}')
plt.legend()
plt.grid(True)
plt.show()
```
