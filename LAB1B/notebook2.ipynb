```python
import numpy as np
import matplotlib.pyplot as plt 

import datagen as dg
import ml_perceptron as mlp
```

```python
classA, classB = dg.generate_data(ndata=300)
dg.visualize_data(classA, classB)
classA.shape
```

```python
# shuffle the samples
data, label = dg.shuffle_data(classA, classB)
data.shape
```

```python
input_dim = 2
hidden_dim = 6
#output_dim = 1

perceptron = mlp.MultiLayersPerceptron(input_dim, hidden_dim, task=mlp.TASK1)
perceptron.train(data, label, epochs=100, lr = 0.01, batch_size=50)
```
